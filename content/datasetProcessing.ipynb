```python
import pandas as p
import numpy as n
import librosa


#Loading and reading the dataset
dataset_file =  "MagnaTagATune_dataset.csv"
x = p.read_csv(dataset_file, delimiter="\t")

#displaying first ten rows
x.head(10)

```

```python
#Extracting Audio features like tempo, spectral centroid and Mel-frequency cepstral coefficients (MFCCs).
def extract_features(file_name): # NOTE here file_name is expected to be a path to an audio , I instaed have a dataset "MagnaAtune "
    try:                            #how does that one factor in all this

        #EXTRACTING AUDIO TIME SERIES AND SAMPLING RATE
        audio, sample_rate = librosa.load(file_name, res_type="kaiser_fast")
        #librosa.load(),  loads the audio file specified by file_name.
        #res_type, the resampling type. 'kaiser_fast' is a high-quality, efficient resampling algorithm.
        #Returns: A tuple containing the audio time series (audio) and the sampling rate (sample_rate).

        #EXTRACTING MFCCs
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        #librosa.feature.mfcc(), extracts MFCCs from the audio signal, y; the audio time series, sr; sampling rate of the audio and n_mfcc; number of mfccs to return
        #Returns: An array of MFCCS
        mfccs_scaled = n.mean(mfccs.T, axis=0)
        #n.mean(), calculates the mean of the MFCCs across teh time axis
        #mfccs.T, Transppses the mfcc aray so that the time dimension is along the columns
        #axis=0 specifies that mean should be calculated along the rows
        #mfccs_scaled, the mean of the MFCCs for each coeficcient

        #EXTRACTING ADDITIONAL FEATURES
        tempo, _ = librosa.beat.beat_track(y=audio, sr=sample_rate)
        #librosa.beat.beat_track() , estimate sthe tempo 9beats per minute) of teh audio
        #Returns: A tuple  containing the estimated tempo and beat frames(not used here, hence '_' )
        spectral_centroid = n.mean(librosa.feature.spectral_centroid(y=audio, sr=sample_rate))
        #librosa.feature.spectral_centroid() indictes where the center of mass of the spectrum is located .

        #RETURNING EXTRACTED FEATURES
        return n.hstack(mfccs_scaled, tempo, spectral_centroid)
         #n.hstack(), horizontally  stack the extracted features into a single array

    except Exception as e:
        print(f"Erro encountered while parssing file: {file_name}")
        return None
```
